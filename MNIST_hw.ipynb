```python
from keras.datasets import mnist
from keras import models
from keras import layers
from keras.utils import to_categorical
from keras import metrics
import numpy as np

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 神經網路設定
network = models.Sequential()
network.add(layers.Dense(1000, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy', 'precision', 'recall', 'f1_score'])

# 資料處理
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# 訓練模型
history = network.fit(train_images, train_labels, epochs=5, batch_size=128, validation_data=(test_images, test_labels))

# 顯示最終績效

# 取得所有測試集績效指標
test_results = network.evaluate(test_images, test_labels)

# 列印測試集績效指標
print('測試集績效')
print("Loss:", test_results[0])
print("accuracy:", test_results[1])
print("precision:", np.mean(test_results[2]))
print("recall:", test_results[3])
print("f1_score:", test_results[4])
print('===================================')
# 顯示訓練集的績效
print('訓練集績效')
train_metrics = history.history
print("accuracy:", train_metrics['accuracy'][-1])
print("loss:", train_metrics['loss'][-1])
print("precision:", train_metrics['precision'][-1])
print("recall:", train_metrics['recall'][-1])

# 計算訓練集的最後一個F1分數
train_precision = train_metrics['precision'][-1]
train_recall = train_metrics['recall'][-1]
train_f1_score = 2 * ((train_precision * train_recall) / (train_precision + train_recall + 1e-10))
print("f1_score:", train_f1_score)
#print("f1_score:", np.mean(train_metrics['f1_score']))
#print("f1_score:", train_metrics['f1_score'][-1])

```
